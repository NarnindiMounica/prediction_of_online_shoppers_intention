### Encoding Categorical Features to Nominal Features

```python
## Target feature
data['Revenue'].value_counts()
```

```python
data['Revenue'] = np.where(data['Revenue']==True, 1,0 )
data['Revenue'].value_counts()
## True as 1 and False as 0
```

```python
## Encoding categorical independent features
from sklearn.preprocessing import LabelEncoder
categorical_features = ['Month', 'VisitorType', 'Weekend']
label_encoder = LabelEncoder()
for i in categorical_features:
    data[i] = label_encoder.fit_transform(data[i])
```

### Splitting Data into Train and Test Sets

```python
x = data[[i for i in data.columns if i != 'Revenue']]
y = data[['Revenue']]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=1)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

### Over-Sampling using SMOTE

Here we have an imbalanced dataset where minority class is Revenue = True/1.
Hence we use over-sampling technique(SMOTE) to synthetically reproduce minority class.
Sampling is done in training set, to eliminate risk of having synthetic data in test set, which was never trained by model.

```python
y_train.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()
x_train_smote, y_train_smote = smote.fit_resample(x_train, y_train)
y_train_smote.value_counts()
```

### Feature Scaling Using Standardization

```python
from sklearn.preprocessing import StandardScaler
standard_scaler = StandardScaler()

x_train_smote_ss =standard_scaler.fit_transform(x_train_smote)
x_test_ss = standard_scaler.transform(x_test)
x_train_smote_ss = pd.DataFrame(x_train_smote_ss, columns=x_train.columns)
x_test_ss = pd.DataFrame(x_test_ss, columns=x_train.columns) 
```

```python
x_train_smote_ss.head()
```

### Model Building and Selection of Best Model

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
import xgboost 
from xgboost import XGBClassifier
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, plot_confusion_matrix, roc_curve, auc
```

```python
k_fold = StratifiedKFold(n_splits = 7, shuffle=True, random_state=1)
```

```python
cv_models= []
cv_models_score = []
cv_models_mean= []
cv_models_std = []
rs = 1

cv_models.append(LogisticRegression(random_state=rs))
cv_models.append(KNeighborsClassifier())
cv_models.append(DecisionTreeClassifier(random_state=rs))
cv_models.append(RandomForestClassifier(random_state=rs))
cv_models.append(AdaBoostClassifier(random_state=rs))
cv_models.append(GradientBoostingClassifier(random_state=rs))
cv_models.append(XGBClassifier(random_state=rs))

for model in cv_models:
    cv_scores = cross_val_score(model, x_train_smote_ss, y_train_smote, scoring='roc_auc', cv=k_fold) 
    cv_models_score.append(cv_scores)
    
    
for score in cv_models_score:
    cv_models_mean.append(score.mean())
    cv_models_std.append(score.std())
    
    
```

```python
Models_Report = pd.DataFrame(data=[cv_models_mean, cv_models_std], columns=['Logistic Regression', 'KNeighbors Classifier', 'DT Classifier', 'RF Classifier', 'AdaBoost Classifier', 'Gradient Boosting', 'XG Boost'], index=['cross_validation_score', 'cross_validation_error'])
```

```python
Models_Report
```

From the scores obtained Best Model is RandomForest Classifier.

### Hyper-Parameter Tuning of Best Model RandomForest Classifier

```python
## with default hyper-parameters
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(random_state=1)
rfc.get_params()
```

```python
rfc_dmodel= rfc.fit(x_train_smote_ss, y_train_smote)
rfc_dmodel_y = rfc_dmodel.predict(x_test_ss)
print(classification_report(y_test, rfc_dmodel_y))
```

```python
rfc_dmodel_auc = roc_auc_score(y_test, rfc_dmodel_y)
print('AUC Score of Random Forest Model with Default Hyper-Parameters: ',round(rfc_dmodel_auc,2))
```

```python
## Hyper-parameter tuning using GridSearchCV

parameters= {'criterion':['entropy','gini'], 'max_features': ['auto', 'sqrt', 'log2', None],  'max_depth': range(4,12,2), 'min_samples_leaf': range(10,50,5) }

from sklearn.model_selection import GridSearchCV
grid_cv = GridSearchCV(rfc, param_grid=parameters, scoring='roc_auc', cv=k_fold)
Grid_RFC = grid_cv.fit(x_train_smote_ss, y_train_smote)
```

```python
Grid_RFC.best_score_
```

```python
Grid_RFC.best_params_
```

```python
rfc.set_params(criterion= 'entropy', max_depth=10, max_features= None, bootstrap=True, min_samples_leaf=10, random_state=1)
                 
```

```python
## OOB Score for different estimators

rfc.set_params(warm_start=False, 
                  oob_score=True)

min_estimators = 100
max_estimators = 500

error_rate = {}

for i in range(min_estimators, max_estimators + 1):
    rfc.set_params(n_estimators=i)
    rfc.fit(x_train_smote_ss, y_train_smote)

    oob_error = 1 - rfc.oob_score_
    error_rate[i] = oob_error
```

```python
# Convert dictionary to a pandas series for easy plotting 
oob_series = pd.Series(error_rate)
```

```python
plt.figure(figsize=(15,6))
plt.plot(oob_series, color='blue')
plt.title('OOB Error for different Estimators')
plt.show()
```

```python
## Here we select n_estimators as 130, since OOB Error is stable and minimum at that value

rfc.set_params(criterion= 'entropy', max_depth=10, n_estimators=130, max_features= None, oob_score=False, bootstrap=True, min_samples_leaf=10, random_state=1)
```
